```python
# Cell 2: Load test cases and run evaluation
import json
from src.llm.chatbot import AI_Waiter

# Load test cases
with open('data/test_cases.json', 'r', encoding='utf-8') as f:
    test_cases = json.load(f)

# Initialize chatbot
print("🔄 Initializing chatbot...")
bot = AI_Waiter()
bot.initialize_system()
print("✅ Chatbot ready!\n")

print("=" * 80)
print(f"🤖 COMPREHENSIVE EVALUATION - {len(test_cases)} TEST CASES")
print("=" * 80)

# Track results
results = []
category_stats = {}

for i, case in enumerate(test_cases, 1):
    question = case['question']
    expected_info = case['expected_info']
    category = case['category']
    
    print(f"\n[{i}/{len(test_cases)}] {category.upper()}")
    print(f"❓ Question: {question}")
    print(f"🎯 Expected: {expected_info}")
    print("-" * 50)
    
    try:
        # Reset history for each test
        bot.chat_history = []
        
        # Get chatbot response
        response = bot.chat(question)
        print(f"🤖 Response:\n{response}")
        
        # Analyze response
        found_keywords = []
        response_lower = response.lower()
        
        for keyword in expected_info:
            if keyword.lower() in response_lower:
                found_keywords.append(keyword)
        
        # Calculate coverage
        coverage = len(found_keywords) / len(expected_info)
        
        # Store results
        result = {
            'question': question,
            'category': category,
            'expected': expected_info,
            'found': found_keywords,
            'coverage': coverage,
            'response_length': len(response)
        }
        results.append(result)
        
        # Update category stats
        if category not in category_stats:
            category_stats[category] = []
        category_stats[category].append(coverage)
        
        print(f"✅ Found: {found_keywords}")
        print(f"📊 Coverage: {coverage:.2f} ({len(found_keywords)}/{len(expected_info)})")
        
    except Exception as e:
        print(f"❌ Error: {e}")
        results.append({
            'question': question,
            'category': category,
            'coverage': 0,
            'error': str(e)
        })
    
    print("=" * 80)

# Final statistics
print(f"\n🎯 EVALUATION SUMMARY")
print("=" * 50)

overall_coverage = sum(r['coverage'] for r in results) / len(results)
print(f"📊 Overall Coverage: {overall_coverage:.3f} ({overall_coverage*100:.1f}%)")

print(f"\n📂 Performance by Category:")
for category, coverages in category_stats.items():
    avg_coverage = sum(coverages) / len(coverages)
    print(f"   {category:12}: {avg_coverage:.3f} ({avg_coverage*100:.1f}%) - {len(coverages)} tests")

# Find best and worst performing categories
if category_stats:
    best_category = max(category_stats.keys(), key=lambda c: sum(category_stats[c])/len(category_stats[c]))
    worst_category = min(category_stats.keys(), key=lambda c: sum(category_stats[c])/len(category_stats[c]))
    
    print(f"\n🏆 Best Category: {best_category}")
    print(f"❌ Worst Category: {worst_category}")

# Save results
with open('data/evaluation_results.json', 'w', encoding='utf-8') as f:
    json.dump({
        'overall_coverage': overall_coverage,
        'category_stats': category_stats,
        'detailed_results': results
    }, f, ensure_ascii=False, indent=2)

print(f"\n💾 Results saved to data/evaluation_results.json")
print("🏁 Evaluation completed!")
```
