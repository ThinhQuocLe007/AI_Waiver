```python
bot = AI_Waiter()
```

```python
bot.initialize_system()
```

```python
print("🍽️  Welcome to AI Waiter!")
print("Type 'exit' or 'quit' to end the conversation\n")

while True:
    user_input = input("👤 You: ").strip()
    
    if user_input.lower() in ['exit', 'quit', 'bye', 'tạm biệt']:
        print("🤖 Waiter: Cảm ơn bạn! Hẹn gặp lại!")
        break
    
    if not user_input:
        print("Please enter something...\n")
        continue
    
    try:
        response = bot.chat(user_input)
    except Exception as e:
        print(f"❌ Error: {e}\n")
```
