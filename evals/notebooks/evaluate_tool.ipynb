```python
# Cell: Run Tool Usage Evaluation
from src.llm.chatbot import AI_Waiter
import json
import re

# Initialize chatbot
print("🔄 Initializing chatbot...")
bot = AI_Waiter()
bot.initialize_system()
print("✅ Chatbot ready!\n")

print("=" * 80)
print("🛠️ TOOL USAGE EVALUATION")
print("=" * 80)

# Track results
results = []
tool_stats = {
    "no_tool": {"correct": 0, "total": 0},
    "search": {"correct": 0, "total": 0},
    "place_order": {"correct": 0, "total": 0}
}

def detect_actual_tool_used(response_log):
    """Detect which tool was actually used based on logs"""
    if "[Search]" in response_log or "search" in response_log.lower():
        return "search"
    elif "[Order]" in response_log or "place_order" in response_log.lower():
        return "place_order"
    else:
        return "no_tool"

for i, case in enumerate(tool_test_cases, 1):
    question = case['question']
    expected_tool = case['expected_tool']
    category = case['category']
    
    print(f"\n[{i}/{len(tool_test_cases)}] {expected_tool.upper()}")
    print(f"❓ Question: {question}")
    print(f"🎯 Expected Tool: {expected_tool}")
    print("-" * 50)
    
    try:
        # Reset history
        bot.chat_history = []
        
        # Capture response (you might need to modify this based on your logging)
        response = bot.chat(question)
        
        # Simple detection - check if response contains search indicators
        actual_tool = "no_tool"  # default
        
        # Check for search patterns in response
        search_patterns = ["Dựa trên thông tin", "menu", "giá", "có", "không"]
        if any(pattern.lower() in response.lower() for pattern in search_patterns) and len(response) > 100:
            actual_tool = "search"
        
        # Check for order patterns
        order_patterns = ["đặt", "gọi", "order", "món"]
        if any(pattern in question.lower() for pattern in order_patterns):
            if expected_tool == "place_order":
                actual_tool = "place_order"
        
        # If it's a simple greeting/thanks and short response
        if len(response) < 50 and expected_tool == "no_tool":
            actual_tool = "no_tool"
            
        # Check if tool was used correctly
        is_correct = (actual_tool == expected_tool)
        
        # Update stats
        tool_stats[expected_tool]["total"] += 1
        if is_correct:
            tool_stats[expected_tool]["correct"] += 1
        
        # Store result
        result = {
            'question': question,
            'expected_tool': expected_tool,
            'actual_tool': actual_tool,
            'is_correct': is_correct,
            'category': category,
            'response_length': len(response)
        }
        results.append(result)
        
        print(f"🤖 Response: {response[:150]}{'...' if len(response) > 150 else ''}")
        print(f"🔧 Detected Tool: {actual_tool}")
        print(f"{'✅' if is_correct else '❌'} Tool Selection: {'CORRECT' if is_correct else 'INCORRECT'}")
        
    except Exception as e:
        print(f"❌ Error: {e}")
        tool_stats[expected_tool]["total"] += 1
        results.append({
            'question': question,
            'expected_tool': expected_tool,
            'actual_tool': 'error',
            'is_correct': False,
            'error': str(e)
        })
    
    print("=" * 80)

# Calculate overall accuracy
total_correct = sum(stats["correct"] for stats in tool_stats.values())
total_tests = sum(stats["total"] for stats in tool_stats.values())
overall_accuracy = total_correct / total_tests if total_tests > 0 else 0

print(f"\n🎯 TOOL USAGE EVALUATION SUMMARY")
print("=" * 50)
print(f"📊 Overall Accuracy: {overall_accuracy:.3f} ({overall_accuracy*100:.1f}%)")

print(f"\n🔧 Tool-wise Performance:")
for tool, stats in tool_stats.items():
    if stats["total"] > 0:
        accuracy = stats["correct"] / stats["total"]
        print(f"   {tool:12}: {accuracy:.3f} ({accuracy*100:.1f}%) - {stats['correct']}/{stats['total']}")

# Find common mistakes
print(f"\n❌ Incorrect Predictions:")
mistakes = [r for r in results if not r.get('is_correct', False)]
for mistake in mistakes[:5]:  # Show first 5 mistakes
    print(f"   Q: {mistake['question'][:40]}...")
    print(f"      Expected: {mistake['expected_tool']} | Got: {mistake.get('actual_tool', 'unknown')}")

# Save results
with open('data/tool_usage_results.json', 'w', encoding='utf-8') as f:
    json.dump({
        'overall_accuracy': overall_accuracy,
        'tool_stats': tool_stats,
        'detailed_results': results
    }, f, ensure_ascii=False, indent=2)

print(f"\n💾 Results saved to data/tool_usage_results.json")
print("🏁 Tool usage evaluation completed!")
```
